```python
regions = []
dataDict = {}

for i,j in dfUtah.iterrows():
    # Region
    lat1 = str(j['Start_Lat'])[0:4]
    lat2 = '1'
    try:
        if str(j['Start_Lat'])[4] in ['0','1']:
            lat2 = '1'
        elif str(j['Start_Lat'])[4] in ['2','3']:
            lat2 = '3'
        elif str(j['Start_Lat'])[4] in ['4','5']:
            lat2 = '5'
        elif str(j['Start_Lat'])[4] in ['6','7']:
            lat2 = '7'
        elif str(j['Start_Lat'])[4] in ['8','9']:
            lat2 = '9'
    except:
        lat2 = '1'
    latFull = lat1 + lat2
    lon1 = str(j['Start_Lng']).replace('-','')[0:5]
    lon2 = '1'
    try:
        if str(j['Start_Lng'])[5] in ['0','1']:
            lon2 = '1'
        elif str(j['Start_Lng'])[5] in ['2','3']:
            lon2 = '3'
        elif str(j['Start_Lng'])[5] in ['4','5']:
            lon2 = '5'
        elif str(j['Start_Lng'])[5] in ['6','7']:
            lon2 = '7'
        elif str(j['Start_Lng'])[5] in ['8','9']:
            lon2 = '9'
    except:
        lon2 = '1'
    lonFull = lon1 + lon2
    region = latFull + ',-' + lonFull
    
    #temperature
    tempF = ''
    if j['Temperature(F)']  <= 32:
        tempF = '<= 32'
    elif j['Temperature(F)'] < 80:
        tempF = '< 80'
    else:
        tempF = '80+'
    
    #dew point
    tempC = (j['Temperature(F)'] - 32) * (5/9)
    Td = tempC - ((100-j['Humidity(%)'])/5)
    dewPoint = ''
    if Td <= 55:
        dewPoint = 'dry'
    elif Td <= 65:
        dewPoint = 'moderate'
    else:
        dewPoint = 'humid'
        
    #pressure
    press = ''
    if j['Pressure(in)'] < 27.3:
        press = 'low'
    elif j['Pressure(in)'] <= 30:
        press = 'normal'
    else:
        press = 'high'
    
    #day/night
    dayNight = j['Sunrise_Sunset']
    
    #precipitation
    precip = ''
    if j['Precipitation(in)'] > 0:
        precip = 'True'
    else:
        precip = 'False'
        
    #traffic signal
    traffSig = j['Traffic_Signal']
    
    #roundabout
    roundA = j['Roundabout']
    
    #stop
    stop = j['Stop']
    
    #traffic final
    traffic = ''
    if str(j['Roundabout']) == 'True':
        traffic = 'True'
    elif str(j['Traffic_Signal']) == 'True':
        traffic = 'True'
    elif str(j['Stop']) == 'True':
        traffic = 'True'
    else:
        traffic = 'False'
    
    #severity
    severity = j['Severity']
    
    #temp
    tempList = [region, tempF, str(dayNight), str(precip)
                #, str(traffSig), str(roundA), str(stop)
                ,traffic]
    #tempList = [tempF,dewPoint]
    
    if region not in regions:
        regions.append(region)
        
    key = ';'.join(tempList)
    
    if key not in dataDict.keys():
        dataDict[key] = {'severity': severity, 'count': 1}
    else:
        dataDict[key]['severity'] += severity
        dataDict[key]['count'] += 1
```

```python
temps = ['<= 32', '< 80', '80+']
dewPoints = ['dry','moderate','humid']
dayNights = ['Day','Night']
precipitations = ['True','False']
traffics = ['True','False']
combinations = [regions,temps,dayNights,precipitations,traffics]
#combinations = [regions,temps,dewPoints,dayNights,precipitations,traffics]
#combinations = [temps,dewPoints]
combos = list(itertools.product(*combinations))

for i in combos:
    text = ''
    for ii in i:
        if text == '':
            text = ii
        else:
            text = text + ';' + ii
    if text not in dataDict.keys():
        dataDict[text] = {'severity': 0, 'count': 0}
print(len(dataDict.keys()))
```

```python
finalRegionList = []
for i in dataDict.keys():
    tempList = i.split(';')
    region = tempList[0]
    regionSplit = region.split(',')
    lat = regionSplit[0]
    lon = regionSplit[1]
    tempF = tempList[1]
    #dewPoint = tempList[2]
    dayNight = tempList[2]
    precip = tempList[3]
    traffic = tempList[4]
    severity = dataDict[i]['severity']
    count = dataDict[i]['count']
    #temp = [region,lat,lon,tempF,dewPoint,dayNight,precip,traffic,severity,count]
    temp = [region,lat,lon,tempF,dayNight,precip,traffic,severity,count]
    finalRegionList.append(temp)
    
#finalRegionDf = pd.DataFrame(finalRegionList, columns = ['region','lat','lon','temperature','dewPoint','dayNight','precip','traffic','severity','cnt'])

finalRegionDf = pd.DataFrame(finalRegionList, columns = ['region','lat','lon','temperature','dayNight','precip','traffic','severity','cnt'])

print(finalRegionDf.head(5))
```
